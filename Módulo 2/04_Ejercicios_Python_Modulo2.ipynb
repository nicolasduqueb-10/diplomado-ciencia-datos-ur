### **Ejercicio 5**
Pide n√∫meros al usuario hasta que escriba 'salir'. Usa `while` y calcula el promedio.

‚úÖ **Ejemplo de entrada y salida esperada**

```plaintext
N√∫mero: 5
N√∫mero: 10
N√∫mero: salir

Promedio: 7.5
```

````python
def promedio():
    numeros = []
    while True:
        entrada = input("N√∫mero: ")
        if entrada.lower() == "salir":
            break
        numeros.append(float(entrada))
    if numeros:
        print("\nPromedio:", sum(numeros)/len(numeros))

promedio()
````

### **Ejercicio 6**
Crea una tupla y muestra el mayor valor usando un ciclo `for`.

‚úÖ **Ejemplo de entrada y salida esperada**

```plaintext
Tupla: (3, 7, 2, 9, 4)

Mayor: 9
```

````python
def mayor_tupla():
    tupla = (3, 7, 2, 9, 4)
    mayor = tupla[0]
    for num in tupla:
        if num > mayor:
            mayor = num
    print("Tupla:", tupla)
    print("\nMayor:", mayor)

mayor_tupla()
````

### **Ejercicio 8**
Crea un diccionario con nombres y edades e imprime los mayores de 18.

‚úÖ **Ejemplo de entrada y salida esperada**

```plaintext
Mayores de edad:
Ana - 22
```

````python
def mayores_edad():
    personas = {"Ana": 22, "Luis": 17, "Pedro": 19}
    print("Mayores de edad:")
    for nombre, edad in personas.items():
        if edad >= 18:
            print(nombre, "-", edad)

mayores_edad()
````

### **Ejercicio 11**
Cuenta cu√°ntas vocales tiene una frase.

‚úÖ **Ejemplo de entrada y salida esperada**

```plaintext
Frase: Hola mundo

Cantidad de vocales: 4
```

````python
def contar_vocales():
    frase = input("Frase: ")
    vocales = "aeiouAEIOU"
    contador = 0
    for letra in frase:
        if letra in vocales:
            contador += 1
    print("\nCantidad de vocales:", contador)

contar_vocales()
````

### **Ejercicio 12**
Simula un cajero autom√°tico con saldo inicial.

‚úÖ **Ejemplo de entrada y salida esperada**

```plaintext
Saldo actual: 1000
Monto a retirar: 200

Retiro exitoso üíµ
```

````python
def cajero():
    saldo = 1000
    try:
        print("Saldo actual:", saldo)
        retiro = float(input("Monto a retirar: "))
        if retiro <= saldo:
            saldo -= retiro
            print("\nRetiro exitoso üíµ")
        else:
            print("\nFondos insuficientes.")
    except:
        print("\nError en la operaci√≥n.")

cajero()
````

### **Ejercicio 14**
Elimina duplicados y ordena manualmente una lista.

‚úÖ **Ejemplo de entrada y salida esperada**

```plaintext
Lista original: [3, 1, 2, 3]

Lista sin duplicados y ordenada: [1, 2, 3]
```

````python
def ordenar_manual():
    lista = [3, 1, 2, 3]
    sin_duplicados = []
    for num in lista:
        if num not in sin_duplicados:
            sin_duplicados.append(num)
    for i in range(len(sin_duplicados)):
        for j in range(i+1, len(sin_duplicados)):
            if sin_duplicados[i] > sin_duplicados[j]:
                sin_duplicados[i], sin_duplicados[j] = sin_duplicados[j], sin_duplicados[i]
    print("Lista original:", lista)
    print("\nLista sin duplicados y ordenada:", sin_duplicados)

ordenar_manual()
````

### **Ejercicio 15**
Filtra filas mayores al promedio y guarda en nuevo CSV.

‚úÖ **Ejemplo de entrada y salida esperada**

```plaintext
Archivo filtrado guardado correctamente üìÅ
```

````python
def filtrar_csv():
    import pandas as pd
    df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
    promedio = df["sepal_length"].mean()
    filtrado = df[df["sepal_length"] > promedio]
    filtrado.to_csv("iris_filtrado.csv", index=False)
    print("Archivo filtrado guardado correctamente üìÅ")

filtrar_csv()
````
